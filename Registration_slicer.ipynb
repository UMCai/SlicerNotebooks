```python
import JupyterNotebooksLib as slicernb
import slicer
import SimpleITK as sitk
import sitkUtils as su
import os

# Set viewer size to 50% of screen size
slicernb.AppWindow.setWindowSize(scale=1.0)
# Hide patient information from slice view
slicernb.showSliceViewAnnotations(False)

data_dir = r'C:\Users\p70089067\Desktop\Registration test CVI03'
os.path.exists(data_dir)
# Clear scene
slicer.mrmlScene.Clear(False)


```

```python
# Load 3D image using SimpleITK
reader1 = sitk.ImageFileReader()
reader1.SetFileName(os.path.join(data_dir, 'Pre-op CT.nrrd'))
image1 = reader1.Execute()
reader2 = sitk.ImageFileReader()
reader2.SetFileName(os.path.join(data_dir, 'Intra-op 2 - zonder filter.nrrd'))
image2 = reader2.Execute()
```

```python
volumeNode = su.PushVolumeToSlicer(image1)
# Prevent automatic brightness/contrast update to make the processed images easier to compare
volumeNode.GetDisplayNode().SetAutoWindowLevel(False)

# Set up view layout
slicer.app.layoutManager().setLayout(slicer.vtkMRMLLayoutNode.SlicerLayoutFourUpView)
# Set up slice views
slicer.util.setSliceViewerLayers(background=volumeNode, fit=True)
# Show volume rendering
slicernb.showVolumeRendering(volumeNode, presetName='MR-Default')
slicer.util.resetThreeDViews()
# Display views
slicernb.ViewDisplay()


```
